(a) For this question, use the extracted data set you created in the preprocessing step ('air_pollution_2.csv'). Use a scatter diagram to find the relation between the 'Traffic_Volume' and 'Particulate_Matter' columns.

```python
# Your code
# Read data from air_pollution_2.csv file as 'df'
df = pd.read_csv('air_pollution_2.csv')

# Read the column 'Traffic_Volume' from dataframe 'df'
Traffic_Volume = df['Traffic_Volume'].values

# Read the column 'Particulate_Matter' from dataframe 'df'
Particulate_Matter= df['Particulate_Matter'].values

# Plot the scatter diagram 
plt.rcParams['figure.figsize'] = (17, 10)
plt.scatter(Traffic_Volume, Particulate_Matter, c='green', s =10)
plt.title('Scatter Plot between Traffic_Volume & Particulate_Matter')

# Relation between Traffic_Volume and Particulate_Matter
print("The relation of Traffic_Volume and Particulate_Matter shows that the data are distributed in 6 clusters")
```

(b) From the previous question you found the relationship between two columns which are suitable for applying clustering methods. Find the two arrays corresponding to these to columns and drop nan values from them. Apply the k_means method for clustering the two mentioned columns. Use three different number of clusters for classifying.

```python
# Your code
from sklearn.cluster import KMeans

# Filter the Traffic_Volume and Particulate_Matter column values and remove the 'nan' values
Traffic_Volume = [i for i in Traffic_Volume if str(i)!="nan"]
Particulate_Matter = [i for i in Particulate_Matter if str(i)!="nan"]

# create array list using numpy 
# and compress the Traffic_Volume and Particulate_Matter values in list
data_for_kMean = np.array(list(zip(Traffic_Volume, Particulate_Matter)))

# EXPLAINATION :
# With reference to the previous question, 
# the relation between Traffic_Volume and Particulate_Matter have 6 clusters
# so the optimized number_of_clusters are 6
# lets apply the K-mean algorithm with three different values of number of clusters (k = 6 , k > 6 , k < 6)

# ---------------------------------------------------------------------------------------------------------------------
# (K = 6)
# When Number of Clusters is 6 
cluster_num = 6
kmeans = KMeans(cluster_num).fit(data_for_kMean)
clusters = kmeans.labels_
centroids = kmeans.cluster_centers_

fig, plots = plt.subplots()
# Color list for plotting 
colors = ['red', 'green', 'blue', 'yellow', 'maroon', 'black', 'purple', 'silver', 'olive', 'teal', 'aqua', 'lime', 'fuchsia']
for cluster_index in range(cluster_num) :
    sub_set = np.array([data_for_kMean[i] for i in range(len(data_for_kMean)) if clusters[i] == cluster_index])
    if len(sub_set) == 0 :
        continue
    plots.scatter(sub_set[:,0], sub_set[:,1], s = 10, c = colors[cluster_index])
plots.scatter(centroids[:,0], centroids[:,1], marker = '*', s = 300, c = 'k')
plt.title('K-Mean with 6 Clusters (k = 6)')

# ---------------------------------------------------------------------------------------------------------------------
# (K < 6)
# When Number of Clusters is 4

cluster_num = 4
kmeans = KMeans(cluster_num).fit(data_for_kMean)
clusters = kmeans.labels_
centroids = kmeans.cluster_centers_

fig, plots = plt.subplots()

for cluster_index in range(cluster_num) :
    sub_set = np.array([data_for_kMean[i] for i in range(len(data_for_kMean)) if clusters[i] == cluster_index])
    if len(sub_set) == 0 :
        continue
    plots.scatter(sub_set[:,0], sub_set[:,1], s = 10, c = colors[cluster_index])
plots.scatter(centroids[:,0], centroids[:,1], marker = '*', s = 300, c = 'k')
plt.title('K-Mean with 4 Clusters (k < 6)')

# ---------------------------------------------------------------------------------------------------------------------
# (k > 6)
# When Number of Clusters is 8

cluster_num = 8
kmeans = KMeans(cluster_num).fit(data_for_kMean)
clusters = kmeans.labels_
centroids = kmeans.cluster_centers_

fig, plots = plt.subplots()
for cluster_index in range(cluster_num) :
    sub_set = np.array([data_for_kMean[i] for i in range(len(data_for_kMean)) if clusters[i] == cluster_index])
    if len(sub_set) == 0 :
        continue
    plots.scatter(sub_set[:,0], sub_set[:,1], s = 10, c = colors[cluster_index])
plots.scatter(centroids[:,0], centroids[:,1], marker = '*', s = 300, c = 'k')
plt.title('K-Mean with 8 Clusters (k > 6)')
```

Among the datasets given for this assignment you will find the files "pg_train" and "pg_test". These two files contain a labeled corpus, already splitted in training and test set.

(a) Perform preprocessing on the training corpus (all lowercase, no punctuation, tokenization, stemming, stopword removal) and obtain a binary document-term matrix; train a logistic classifier with the author as target.

```python
# Your code
from sklearn import linear_model, decomposition, datasets
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
import csv

# read data from csv file for training dataset
pg_train = []

with open("pg_train.csv", encoding="ISO-8859-1") as csvFile:
    reader = csv.reader(csvFile, delimiter=';')
    for row in reader:
        pg_train.append(row)
        
        
# read data from csv file for training dataset
pg_test = []

with open("pg_test.csv", encoding="ISO-8859-1") as csvFile:
    reader = csv.reader(csvFile, delimiter=';')
    for row in reader:
        pg_test.append(row)

pg_train_data = []
pg_train_target = []

# separate the descriptive(author data) & target(author name) feature of training dataset using split function 
# the author name and author data is separated by '#' symbol
for row in pg_train:
    if (len(row) != 0):
        temp = ''.join(row).split('#')
        pg_train_target.append(temp[0])
        pg_train_data.append(temp[1])
        
pg_test_data = []
pg_test_target = []

# separate the descriptive(author data) & target(author name) feature of testing dataset using split function 
# the author name and author data is separated by '#' symbol
for row in pg_test:
    if (len(row) != 0):
        temp = ''.join(row).split('#')
        pg_test_target.append(temp[0])
        pg_test_data.append(temp[1])
```

```python
# Train a model using Logistic Classifier with binary document term matrix

logistic = linear_model.LogisticRegression(max_iter=1000)
text_clf = Pipeline([('vect', CountVectorizer(binary=True)), ('logistic', logistic)])
model_1 = text_clf.fit(pg_train_data, pg_train_target)
```

(d) Perform preprocessing on the training corpus (all lowercase, no punctuation, tokenization, stemming, stopword removal) and obtain a doc2vec embedding in order to reduce the dimension of the document vector to 300; use the doc2vec model you just trained to convert the training set to a set of document vectors; train a logistic classifier with the author as target.


```python
# Your code
import gensim
import multiprocessing
from sklearn import utils
from gensim.models import Doc2Vec
from tqdm import tqdm
from sklearn.linear_model import LogisticRegression

# List used to hold the tagged document for training and testing dataset
train_tagged = []
test_tagged = []

# Tokenizing, normalizing, and creating lists of TaggedDocument objects using gensim

for i in range (0, len(pg_train_data)):
    train_tagged.append(gensim.models.doc2vec.TaggedDocument(words=gensim.utils.simple_preprocess(pg_train_data[i]), tags=pg_train_target[i]))

for i in range (0, len(pg_test_data)):
    test_tagged.append(gensim.models.doc2vec.TaggedDocument(words=gensim.utils.simple_preprocess(pg_test_data[i]), tags=pg_test_target[i]))

cores = multiprocessing.cpu_count()

# obtain a doc2vec embedding in order to reduce the dimension of the document vector to 300
doc2vec_model = Doc2Vec(dm=0, vector_size=300, workers=cores)
doc2vec_model.build_vocab([x for x in tqdm(train_tagged)])

# Training the doc2vec model
for epoch in range(30):
    doc2vec_model.train(utils.shuffle([x for x in tqdm(train_tagged)]), total_examples=len(train_tagged), epochs=1)
    doc2vec_model.alpha -= 0.002
    doc2vec_model.min_alpha = doc2vec_model.alpha
    
# Building the feature vector for the classifier
def vec_for_learning(model, docs):
    doc2vec_vectors = [model.infer_vector(doc.words) for doc in docs]
    targets = [doc.tags[0] for doc in docs]
    return doc2vec_vectors, targets

# Translating docs into vectors for training and test set
X_train, y_train = vec_for_learning(doc2vec_model, train_tagged)
X_test, y_test = vec_for_learning(doc2vec_model, test_tagged)

# Train a model using Logistic Classifier
log_reg = LogisticRegression(n_jobs=1, C=1e5,max_iter=10000)
model_4 = log_reg.fit(X_train, y_train)
```
